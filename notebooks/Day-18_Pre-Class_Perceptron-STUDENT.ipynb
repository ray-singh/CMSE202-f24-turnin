<font size=+3>&#9998;</font> It must be linearly seperable.

&#9989; **Question 3**: How is this approach similar to the logistic regression classifier we used previously? How is it different?

<font size=+3>&#9998;</font> Both the Perceptron and Logistic Regression classifiers find a linear boundary to separate classes and update weights iteratively. However, the Perceptron uses a hard threshold and only works for linearly separable data, while Logistic Regression uses a sigmoid function to output probabilities, allowing it to handle some overlap between classes. Logistic Regression is also more interpretable, as it provides confidence levels through probabilities.

&#9989; **Question 4**: What is the update rule used to improve the weights of the neurons?


<font size=+3>&#9998;</font> Learning rate.

&#9989; **Question 5**: What role does the "learning rate" play in the Perceptron model? How does changing the learning rate impact the behavior of the perceptron model works?

<font size=+3>&#9998;</font> If the learning rate is too high, the model may make large jumps in weight values, potentially overshooting the optimal solution, leading to instability in the training process. If the learning rate is too low, the model will make very small updates, causing it to converge slowly, which can increase the training time.
